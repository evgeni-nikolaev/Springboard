# Tire 1: SQLite3

```python
from sqlite3 import Error
```

```python
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn
```

```python
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT *
        FROM FACILITIES
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)
```

```python
def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```

__Q7: Produce a list of all members who have used a tennis court.
Include in your output the name of the court, and the name of the member
formatted as a single column. Ensure no duplicate data, and order by
the member name.__

```python
answer_Q7 = pd.read_sql_query("SELECT DISTINCT sub.court, sub.firstname,  sub.surname \
                              FROM ( \
                                  SELECT Facilities.name AS court, Members.firstname AS firstname, Members.surname AS surname \
                                  FROM Bookings \
                                  INNER JOIN Facilities ON Bookings.facid = Facilities.facid \
                                  AND Facilities.name LIKE  'Tennis Court%' \
                                  INNER JOIN Members ON Bookings.memid = Members.memid) sub \
                              GROUP BY sub.court, sub.firstname, sub.surname \
                              ORDER BY sub.firstname, sub.surname;", 
                              engine)
answer_Q7['name'] = (answer_Q7['firstname'] + ' ' + answer_Q7['surname'])
answer_Q7.drop(['firstname', 'surname'], axis=1, inplace=True)
print('The answer to Q7 is:')
answer_Q7
```

__Q8: Produce a list of bookings on the day of 2012-09-14 which
will cost the member (or guest) more than $30. Remember that guests have
different costs to members (the listed costs are per half-hour 'slot'), and
the guest user's ID is always 0. Include in your output the name of the
facility, the name of the member formatted as a single column, and the cost.
Order by descending cost, and do not use any subqueries.__

```python
answer_Q8 = pd.read_sql_query("SELECT Facilities.name AS facility, Members.firstname,   Members.surname, \
                              CASE WHEN Bookings.memid = 0 THEN Facilities.guestcost * Bookings.slots \
                                   ELSE Facilities.membercost * Bookings.slots \
                                   END AS cost \
                              FROM Bookings \
                              INNER JOIN Facilities ON Bookings.facid = Facilities.facid \
                              AND Bookings.starttime LIKE  '2012-09-14%' \
                              AND (((Bookings.memid =0) AND (Facilities.guestcost * Bookings.slots >30)) \
                              OR ((Bookings.memid !=0) AND (Facilities.membercost * Bookings.slots >30))) \
                              INNER JOIN Members ON Bookings.memid = Members.memid \
                              ORDER BY cost DESC;",
                              engine)
answer_Q8['name'] = answer_Q8['firstname'] + ' ' + answer_Q8['surname']
answer_Q8.drop(['firstname', 'surname'], axis=1, inplace=True)
answer_Q8 = answer_Q8[['facility', 'name', 'cost']]
print('The answer to Q8 is:')
answer_Q8
```

__Q9: This time, produce the same result as in Q8, but using a subquery.__

```python
answer_Q9 = pd.read_sql_query("SELECT * \
                               FROM (  SELECT Facilities.name AS facility, Members.firstname, Members.surname, \
                                       CASE WHEN Bookings.memid = 0 THEN Facilities.guestcost * Bookings.slots \
                                       ELSE Facilities.membercost * Bookings.slots \
                                       END AS cost \
                                       FROM Bookings \
                                       INNER JOIN Facilities ON Bookings.facid = Facilities.facid \
                                       AND Bookings.starttime LIKE  '2012-09-14%' \
                                       INNER JOIN Members ON Bookings.memid = Members.memid \
                               )sub \
                               WHERE sub.cost >30 \
                               ORDER BY sub.cost DESC;",
                             engine)
answer_Q9['name'] = answer_Q9['firstname'] + ' ' + answer_Q9['surname']
answer_Q9.drop(['firstname', 'surname'], axis=1, inplace=True)
answer_Q9 = answer_Q9[['facility', 'name', 'cost']]
print('The answer to Q9 is:')
answer_Q9
```

__Q10: Produce a list of facilities with a total revenue less than 1000.
The output of facility name and total revenue, sorted by revenue. Remember
that there's a different cost for guests and members!__

```python
answer_Q10 = pd.read_sql_query("SELECT * \
                                FROM ( SELECT sub.facility, SUM( sub.cost ) AS total_revenue \
                                       FROM (   SELECT Facilities.name AS facility, \
                                                CASE WHEN Bookings.memid = 0 THEN Facilities.guestcost * Bookings.slots \
                                                ELSE Facilities.membercost * Bookings.slots \
                                                END AS cost \
                                                FROM Bookings \
                                                INNER JOIN Facilities ON Bookings.facid = Facilities.facid \
                                                INNER JOIN Members ON Bookings.memid = Members.memid \
                                        )sub \
                                GROUP BY sub.facility \
                                )sub2 \
                                WHERE sub2.total_revenue <1000;",
                               engine)
print('The answer to Q10 is:')
answer_Q10
```

__Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order__

```python
answer_Q11 = pd.read_sql_query("SELECT m.surname, \
                                       m.firstname, \
                                       m.recommendedby AS recomender_id, \
                                       r.surname AS recomender_surname, \
                                       r.firstname AS recomender_firstname \
                                FROM Members AS m \
                                LEFT JOIN Members AS r ON m.recommendedby = r.memid \
                                WHERE m.recommendedby != 0 \
                                ORDER BY m.surname, m.firstname;",
                               engine)
answer_Q11["Member's Name"] = answer_Q11['firstname'] + ' ' + answer_Q11['surname']
answer_Q11.drop(['firstname', 'surname'], axis=1, inplace=True)
answer_Q11["Recomender's Name"] = answer_Q11['recomender_firstname'] + ' ' + answer_Q11['recomender_surname']
answer_Q11.drop(['recomender_firstname', 'recomender_surname', 'recomender_id'], axis=1, inplace=True)
print('The answer to Q11 is:')
answer_Q11
```

__Q12: Find the facilities with their usage by member, but not guests__

```python
answer_Q12 = pd.read_sql_query("SELECT b.facid, COUNT( b.memid ) AS mem_usage, f.name \
                                FROM (  SELECT facid, memid \
                                        FROM Bookings \
                                        WHERE memid != 0 \
                                    ) AS b \
                                LEFT JOIN Facilities AS f ON b.facid = f.facid \
                                GROUP BY b.facid;",
                               engine)
print('The answer to Q12 is:')
answer_Q12
```
